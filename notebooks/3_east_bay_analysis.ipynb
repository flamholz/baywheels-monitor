```python
station_means = station_timeseries_df.groupby('station_id').mean()
cols = ['fraction_full', 'full', 'empty', 'full_or_empty', ]
sorted_means = station_means.sort_values('full', ascending=False).copy()
for i, station_id in enumerate(sorted_means.index):
    print('Station ID', station_id)
    
    my_station = station_df[station_df.station_id == station_id]
    cap = 0
    if my_station.size:
        name = my_station['name'].values[0]
        city = my_station['city'].values[0]
        cap = my_station['capacity'].values[0]
        print(name, city)
        print('capacity', cap)
    
    if cap == 0:
        continue
    
    my_station_means = sorted_means.loc[station_id]
    for c in cols:
        print('\t', c, my_station_means[c])

    if i > 30:
        break
```
